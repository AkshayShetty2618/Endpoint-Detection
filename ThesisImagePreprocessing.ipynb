# Colormasking

```python
import cv2
import numpy as np
import matplotlib.pyplot  as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
from matplotlib.colors import hsv_to_rgb
```

```python
impath = r'F:\TUhh\Thesis\images\image_23.jpg'
img = cv2.imread(impath)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
r, g, b = cv2.split(img)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")
pixel_colors = img.reshape((np.shape(img)[0]*np.shape(img)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
h, s, v = cv2.split(hsv)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")
axis.scatter(h.flatten(), s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Saturation")
axis.set_zlabel("Value")
plt.show()
```

```python
plt.figure()

axis = plt.subplot(211)
axis.scatter(h.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Value")

axis1 = plt.subplot(212)
axis1.scatter(s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis1.set_xlabel("Saturation")
axis1.set_ylabel("Value")

plt.show()
```

```python
light = (0, 50, 0)
dark = (12, 250, 200)
```

```python
m = img.shape[0]
n = img.shape[1]
blank_image = np.ones((m,n,3), np.uint8)
blank_image[:] = (0, 0, 255)
plt.imshow(blank_image)
patch = blank_image
```

```python
mask = cv2.inRange(hsv, light, dark)
m1=cv2.bitwise_not(mask)
```

```python
imagePatch = cv2.bitwise_and(patch,patch,mask = mask)
rst=img
rst = cv2.bitwise_and(rst,rst,mask=m1)
rst = cv2.add(rst,imagePatch,dtype=cv2.CV_8UC1)
rst = rst.astype(np.uint8)
plt.imshow(rst)
plt.show()
```

```python
cv2.imshow('image',rst)
cv2.waitKey(0)
```
